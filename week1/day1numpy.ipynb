```python
import time
```

```python
python_list = list(range(1000000)) 
```

```python
numpy_array = np.array(range(1000000))
```

```python
start = time.time()           
result = []
for number in python_list:
    result.append(number * 2)  # multiply each number by 2 one by one

end = time.time()             # note the end time

python_time = end - start
print("Normal Python time:", round(python_time, 4), "seconds")
```

```python
start = time.time()           # note the start time

result = numpy_array * 2      # multiply ALL numbers at once

end = time.time()             # note the end time

numpy_time = end - start
print("NumPy time:", round(numpy_time, 4), "seconds")

# ------- How much faster? -------
print("NumPy is", round(python_time/numpy_time), "times faster!")
```
