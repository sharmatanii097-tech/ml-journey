But What if You Want SAME Random Numbers Every Time?
Sometimes in ML you want reproducible results:
You train a model and get 95% accuracy
Your friend runs same code and gets 93%
Why different? Random initialization!

You want SAME results every time
So you can compare fairly

```python
# Without seed - different every time
print(np.random.randn(3))
print(np.random.randn(3))
# Different numbers!
```

```python
# With seed - same every time!
np.random.seed(42)
print(np.random.randn(3))
```

```python
np.random.seed(42)
print(np.random.randn(3))
# Same numbers every time!
```

```python
np.random.seed(42)
print(np.random.randn(3))
print()
```

```python
# Masking
print("\n--- Masking ---")
data = np.array([10, 20, 30, 40, 50])
mask = np.array([1, 0, 1, 0, 1])
result = data * mask
print("Original data:", data)
print("Mask:", mask)
print("After masking:", result)
```
