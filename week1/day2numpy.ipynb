```python
import numpy as np
```

```python
# np.zeros - array filled with all zeros
a = np.zeros((3, 4))
print("Zeros array:")
print(a)
print()
```

```python
print("Shape:", a.shape)
print("dtype:", a.dtype)

print()
```

```python
# np.ones - array filled with all ones
b = np.ones((2, 3))
print("Ones array:")
print(b)
print("Shape:", b.shape)
print("dtype:", b.dtype)
```

```python
# arange
c = np.arange(0, 10, 2)
print("Arange array:")
print(c)
print()
print("dtype:", c.dtype)
```

```python
# linspace
d = np.linspace(0, 1, 5)
print("Linspace array:")
print(d)
print()
print("dtype:", d.dtype)
```

```python
# random
e = np.random.randn(3, 3)
print("Random array:")
print(e)
print("Shape:", e.shape)
print("dtype:", e.dtype)
```
